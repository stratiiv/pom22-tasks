## Task 1
Write the programm that calculate total price with discount by the products.

Use class Product(name, price, count) and class Cart. In class Cart you can add the products.

Discount depends on count product:


count	discount

at least 5 -- 5%

at least 7 -- 10%

at least 10 -- 20%

at least 20 -- 30%

more than 20 -- 50%

Write unittest with class CartTest and test all methods with logic

##### **SOLUTION:**

```python
import unittest

class Product:
    def __init__(self,name,price,count):
        self.name=name
        self.price=price
        self.count=count


class Cart:
    def __init__(self,products):
        self.products=products
    @staticmethod
    def after_discount(disc,price):
        return price-price*disc
    def get_total_price(self):
        total_price=0.0
        for product in self.products:
            if product.count<5:
                total_price+=product.price*product.count
            elif 7>product.count>=5:
                total_price += self.after_discount(0.05, product.price*product.count)
            elif 10>product.count>=7:
                total_price += self.after_discount(0.1, product.price*product.count)
            elif 20>product.count >= 10:
                total_price += self.after_discount(0.2, product.price*product.count)
            elif product.count == 20:
                total_price += self.after_discount(0.3, product.price*product.count)
            elif product.count>20 :
                total_price += self.after_discount(0.5, product.price*product.count)
        return total_price

class CartTest(unittest.TestCase):
    def setUp(self):
        products = (Product('p1', 10, 4),
            Product('p2', 100, 5),
            Product('p3', 200, 6),
            Product('p4', 300, 7),
            Product('p5', 400, 9),
            Product('p6', 500, 10),
            Product('p7', 1000, 20))
        cart = Cart(products)
        self.actual=cart.get_total_price()
        self.expected=24785.0
    def test_total_price(self):
        self.assertEqual(self.actual,self.expected)
        
```

## Task 3
Write the function quadratic_equation with arguments a, b ,c that solution to quadratic equation without a complex solution.

Write unit tests for this function with QuadraticEquationTest class

Minimum 3 tests: discriminant < 0, discriminant > 0, discriminant = 0

##### **SOLUTION:**

```python
import unittest
import math

def quadratic_equation(a, b, c):
    if not a and not b and not c:
        raise ValueError
    d=b**2-(4*a*c)
    if d>0:
        return (-b+math.sqrt(d))/(2*a),(-b-math.sqrt(d))/(2*a)
    elif d==0:
        return -b/2*a
    else:
        return None

class QuadraticEquationTest(unittest.TestCase):
    def test_discr_greater_zero(self):
        actual=quadratic_equation(2,1,-1)
        expected=(0.5,-1)
        self.assertEqual(actual,expected)
    def test_discr_equals_zero(self):
        actual=quadratic_equation(1,-4,4)
        expected=2.0
        self.assertEqual(actual,expected)
    def test_discr_less_zero(self):
        actual=quadratic_equation(4,1,2)
        expected=None
        self.assertEqual(actual,expected)
    def test_raises(self):
        with self.assertRaises(ValueError):
            quadratic_equation(0,0,0)
        
```

test data:

##### **SOLUTION:**

```python
import unittest
import math
class TriangleNotValidArgumentException(Exception):
    pass


class TriangleNotExistException(Exception):
    pass


class Triangle:
    def __init__(self,data):
        if type(data) not in (list,tuple) or any(type(i) not in (int,float) for i in data) or len(data)!=3:
            raise TriangleNotValidArgumentException("Not valid arguments")
        a,b,c=data
        if (a + b <= c) or (a + c <= b) or (b + c <= a):
            raise TriangleNotExistException("Can`t create triangle with this arguments")
        self.a=a
        self.b=b
        self.c=c
    def get_area(self):
        p=(self.a+self.b+self.c)/2
        s=math.sqrt(p*(p-self.a)*(p-self.b)*(p-self.c))
        return s

class TriangleTest(unittest.TestCase):
    valid_test_data = [
        ((3, 4, 5), 6.0),
        ((10, 10, 10), 43.30),
        ((6, 7, 8), 20.33),
        ((7, 7, 7), 21.21),
        ((50, 50, 75), 1240.19),
        ((37, 43, 22), 406.99),
        ((26, 25, 3), 36.0),
        ((30, 29, 5), 72.0),
        ((87, 55, 34), 396.0),
        ((120, 109, 13), 396.0),
        ((123, 122, 5), 300.0)
    ]
    not_valid_triangle = [
        (1, 2, 3),
        (1, 1, 2),
        (7, 7, 15),
        (100, 7, 90),
        (17, 18, 35),
        (127, 17, 33),
        (145, 166, 700),
        (1000, 2000, 1),
        (717, 17, 7),
        (0, 7, 7),
        (-7, 7, 7)
    ]
    not_valid_arguments = [
        ('3', 4, 5),
        ('a', 2, 3),
        (7, "str", 7),
        ('1', '1', '1'),
        'string',
        (7, 2),
        (7, 7, 7, 7),
        'str',
        10,
        ('a', 'str', 7)
    ]
    def test_valid_data(self):
        for data in TriangleTest.valid_test_data:
            with self.subTest():
                self.assertAlmostEqual(Triangle(data[0]).get_area(),data[1],1)
    def test_not_valid_triangle(self):
        for data in TriangleTest.not_valid_triangle:
            with self.assertRaises(TriangleNotExistException):
                Triangle(data)
    def test_not_valid_arguments(self):
        for data in TriangleTest.not_valid_arguments:
            with self.assertRaises(TriangleNotValidArgumentException):
                Triangle(data)

```

## Task 5
Create class Worker with fields name and salary. If salary negative raise ValueError

Create a method get_tax_value() that calculate tax from salary . Tax must be calculate like "progressive tax" with next step:

* less then 1000 - 0%
* 1001 - 3000 - 10%
* 3001 - 5000 - 15%
* 5001 - 10000 - 21%
* 10001 - 20000 - 30%
* 20001 - 50000 - 40%
* more than 50000 - 47%

Please create WorkerTest class with unitest to the class Worker. Try to use setUp and tearDown methods. Don`t use assertRaises in tests.

##### **SOLUTION:**

```python
import unittest


class Worker:
    def __init__(self,name,salary=0):
        self.name=name
        if salary<0:
            raise ValueError
        self.salary=salary
    def get_tax_value(self):
        yearly=self.salary
        percents = [0,0.1, 0.15, 0.21, 0.3, 0.4,0.47]
        thresholds = [0, 1000, 3000, 5000, 10000,20000,50000]
        taxes = 0.0
        for i in range(len(percents)):
            if i == len(thresholds) - 1 or yearly < thresholds[i + 1]:
                taxes += (yearly - thresholds[i]) * percents[i]
                break
            else:
                taxes += (thresholds[i + 1] - thresholds[i]) * percents[i]
        return taxes


class WorkerTest(unittest.TestCase):
    def setUp(self) -> None:
        self.data=((Worker("Vika", 100000),40050.0),(Worker("Natasha", 1001),0.1),(Worker("Vasia"),0.0))
    def test_valid(self):
        for worker in self.data:
            with self.subTest():
                self.assertEqual(worker[0].get_tax_value(),worker[1])
    @unittest.expectedFailure
    def test_invalid(self):
        self.assertEqual(self.data[0][0].get_tax_value(),-5)
    def tearDown(self) -> None:
        self.data=None
```

## Task 6
Create function file_parser. If function is called with 2 arguments it must count the number of occurrences string in a file, in case of 3 arguments it must replace string in a file to new string

first argument - path to file

second argument - find string

third argument - replace string

Function must returned string with count of occurrences or count of replaced strings

Example:

file_parser("file.txt", 'x', 'o')-> Replaced 8 strings
file_parser("file.txt", 'o') -> Found 8 strings
Please, create class ParsesTest and write unittest for file_parser function uses mock object

##### **SOLUTION:**

```python
import unittest
from unittest.mock import patch
import io
def file_parser(file,*args):
    with open(file, "rt") as fin:
        with open("out.txt", "wt") as fout:
            if len(args)==2:
                count = fin.read().count(args[0])
                for line in fin:
                    fout.write(line.replace(args[0], args[1]))
                return f'Replaced {count} strings'
            else:
                count=fin.read().count(args[0])
                return f'Found {count} strings'
class ParserTest(unittest.TestCase):
    @patch('builtins.open')
    def test_find(self,mock_open):
        mock_open.return_value=io.StringIO("some initial text data")
        self.assertEqual(file_parser('any.txt','a'),'Found 3 strings')
    @patch('builtins.open')
    def test_replace(self,mock_open):
        mock_open.return_value = io.StringIO("some initial text data")
        self.assertEqual(file_parser('any.txt','i','1'),'Replaced 3 strings')
```
